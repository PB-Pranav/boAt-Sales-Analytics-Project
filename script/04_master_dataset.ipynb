```python
import sys
import pandas as pd
sys.path.append('notebooks/src')
from Data_Loader import load_cleaned

sales, channels, products, customers, delivery = load_cleaned()
print("✅ Data loaded!")
```

```python
# Check how many null dates exist
print("Null OrderDates:", sales['OrderDate'].isna().sum())
print("Total rows:", len(sales))

# See what those rows look like
print(sales[sales['OrderDate'].isna()].head())
```

```python
master = (sales
    .merge(channels,  on='ChannelID',         suffixes=('', '_ch'))
    .merge(products,  on='ProductID',         suffixes=('', '_pr'))
    .merge(customers, on='CustomerID',        suffixes=('', '_cu'))
    .merge(delivery,  on='DeliveryPartnerID', suffixes=('', '_del'))
)

master.to_csv(r'C:\Users\Hello\Downloads\project\BoatSalesProject\data\cleaned\master_sales.csv', index=False)
print(f"✅ Master dataset saved — Shape: {master.shape}")
print(master.head())
```
