```python
import sys
import pandas as pd
sys.path.append('notebooks/src')
from Data_Loader import load_raw

sales, channels, products, customers, delivery = load_raw()
print("✅ Raw data loaded!")
```

```python
# Check what your dates actually look like
print(sales['OrderDate'].head(100000))
print("---")
print(customers['SignupDate'].head(10))
```

```python
sales['OrderDate'] = pd.to_datetime(sales['OrderDate'], format='mixed')
customers['SignupDate'] = pd.to_datetime(customers['SignupDate'], format='mixed')
print("✅ Dates fixed with mixed formatting")
```

```python
sales['Year']      = sales['OrderDate'].dt.year
sales['Month']     = sales['OrderDate'].dt.month
sales['Quarter']   = sales['OrderDate'].dt.quarter
sales['MonthName'] = sales['OrderDate'].dt.strftime('%b')
print(sales[['OrderDate','Year','Month','Quarter','MonthName']].head())
```

```python
#handle nulls
sales['CancellationReason'] = sales['CancellationReason'].fillna('None')
sales['ReturnReason']       = sales['ReturnReason'].fillna('None')
print("✅ Nulls filled")
```

```python
sales['DiscountPct']  = pd.to_numeric(sales['DiscountPct'],  errors='coerce')
sales['DeliveryDays'] = pd.to_numeric(sales['DeliveryDays'], errors='coerce')
print("✅ Data types fixed")
```

```python
#remove duplicate un salesid
before = len(sales)
sales.drop_duplicates(subset='SalesID', inplace=True)
after = len(sales)
print(f"✅ Removed {before - after} duplicate rows")
```

```python
#bussiness logic
assert (sales['NetSales'] >= 0).all(), "❌ Negative NetSales found!"
print("✅ Business logic OK — no negative NetSales")
```
