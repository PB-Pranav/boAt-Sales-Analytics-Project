```python
sales.to_sql('salesdata',           engine, if_exists='replace', index=False)
print("✅ salesdata loaded")

channels.to_sql('saleschannel',     engine, if_exists='replace', index=False)
print("✅ saleschannel loaded")

products.to_sql('productdetails',   engine, if_exists='replace', index=False)
print("✅ productdetails loaded")

customers.to_sql('customerdetails', engine, if_exists='replace', index=False)
print("✅ customerdetails loaded")

delivery.to_sql('deliverypartner',  engine, if_exists='replace', index=False)
print("✅ deliverypartner loaded")

print("\n✅ All 5 tables loaded into MySQL!")
```
